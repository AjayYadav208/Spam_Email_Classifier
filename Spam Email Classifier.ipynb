```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load the dataset
# Assume the dataset is a CSV file with 'text' and 'label' columns
df = pd.read_csv('emails.csv')

# Remove null values and duplicates
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

# Display basic information about the dataset
print(df.info())
print(df.head())
```

```python
# Split the dataset into features and labels
X = df['text']
y = df['spam']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize the TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# Transform the testing data
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Naive Bayes Classifier
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)

# Support Vector Machine Classifier
svm_model = SVC(kernel='linear')
svm_model.fit(X_train_tfidf, y_train)
```

```python
# Predictions with Naive Bayes
nb_predictions = nb_model.predict(X_test_tfidf)

# Predictions with SVM
svm_predictions = svm_model.predict(X_test_tfidf)

# Function to evaluate the model
def evaluate_model(predictions, y_test):
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions)
    recall = recall_score(y_test, predictions)
    f1 = f1_score(y_test, predictions)
    return accuracy, precision, recall, f1

# Evaluate SVM model
svm_accuracy, svm_precision, svm_recall, svm_f1 = evaluate_model(svm_predictions, y_test)

# Print the evaluation metrics
print(f"SVM - Accuracy: {svm_accuracy}, Precision: {svm_precision}, Recall: {svm_recall}, F1 Score: {svm_f1}")
```

```python
# Load a different dataset for testing (assuming it's in the same format)
df_new = pd.read_csv('spam_assassin.csv')

# Preprocess the new dataset
df_new.dropna(inplace=True)
df_new.drop_duplicates(inplace=True)

# Split the new dataset
X_new = df_new['text']
y_new = df_new['target']

# Transform the new data using the same TF-IDF vectorizer
X_new_tfidf = tfidf_vectorizer.transform(X_new)

# Predictions with the new dataset using Naive Bayes and SVM
nb_new_predictions = nb_model.predict(X_new_tfidf)
svm_new_predictions = svm_model.predict(X_new_tfidf)

# Evaluate the models on the new dataset
svm_new_accuracy, svm_new_precision, svm_new_recall, svm_new_f1 = evaluate_model(svm_new_predictions, y_new)

# Print the evaluation metrics for the new dataset
print(f"SVM (New Data) - Accuracy: {svm_new_accuracy}, Precision: {svm_new_precision}, Recall: {svm_new_recall}, F1 Score: {svm_new_f1}")
```
